```r
library(goseq)
suppressPackageStartupMessages(library('GenomicFeatures'))
```

```r
in_dir                     = '../MJ_analysis/WGCNA/'
mapping_filename           = '../data/gene_names_mapping_new.tsv'

modules = c('black', 'blue', 'brown', 'cyan', 'darkgreen', 'darkgrey', 'darkmagenta', 'darkolivegreen', 'darkorange', 'darkred', 'darkturquoise', 'green', 'greenyellow', 'grey60', 'grey', 'lightcyan', 'lightgreen', 'lightyellow', 'magenta', 'midnightblue', 'orange', 'paleturquoise', 'pink', 'purple', 'red', 'royalblue', 'saddlebrown', 'salmon', 'skyblue', 'steelblue', 'tan', 'turquoise', 'violet', 'white', 'yellow')
```

```r
map             = read.table(mapping_filename, sep='\t', header=TRUE, stringsAsFactors = F)
rownames(map)   = map$ensembl
```

```r
all_genes = c()
for (i in 1:length(modules)){
    module = modules[i]
    infile = paste0(in_dir, '16p12_LCL_wgcna_', module, '_module_combat.txt')

    module_genes = scan(infile, what=character())
    module_genes = unlist(lapply(module_genes, function(s) (strsplit(s, '.', fixed=T)[[1]][1])))

    gene_lengths = map[module_genes,'exons_length']
    
    all_genes = c(all_genes, module_genes)
}
gene_lengths = map[all_genes,'exons_length']
```

```r
i = 1
module = modules[i]
infile = paste0(in_dir, '16p12_LCL_wgcna_', module, '_module_combat.txt')

module_genes = scan(infile, what=character())
module_genes = unlist(lapply(module_genes, function(s) (strsplit(s, '.', fixed=T)[[1]][1])))
genes = as.integer(all_genes%in%module_genes)
names(genes) = all_genes
```

```r
for (i in 1:length(modules)) {
    module = modules[i]
    infile = paste0(in_dir, '16p12_LCL_wgcna_', module, '_module_combat.txt')

    module_genes = scan(infile, what=character())
    module_genes = unlist(lapply(module_genes, function(s) (strsplit(s, '.', fixed=T)[[1]][1])))
    genes = as.integer(all_genes%in%module_genes)
    names(genes) = all_genes

    pwf=nullp(genes, bias.data = gene_lengths)
    go=goseq(pwf,"hg19","ensGene")
    go$over_rep_FDR = p.adjust(go$over_represented_pvalue, method="BH")

    go_sig = go[go$over_rep_FDR < .05,]
    print(dim(go_sig))
    print(go_sig$term)

    out_dir = 'wgcna_go/'
    write.table(go_sig, paste0(out_dir, '16p12_LCL_wgcna_', module, '_module_go.tsv'), 
                    sep='\t', row.names=F)
}
```
