```python
import pandas as pd
```

```python
# load in GTEx eQTLs
gtex = pd.read_csv('/data5/16p12_RNA/GTEx/data/GTEx_Analysis_v7_eQTL/Cells_EBV-transformed_lymphocytes.v7.signif_variant_gene_pairs.txt.gz', sep='\t')
```

```python
# create df for annovar input
# has columns:
# chrom	start	end	ref	alt

# get chrom
gtex['chrom'] = gtex.variant_id.apply(lambda s: 'chr' + s.split('_')[0])

# get start
gtex['start'] = gtex.variant_id.apply(lambda s: int(s.split('_')[1]))

# get ref
gtex['ref'] = gtex.variant_id.apply(lambda s: s.split('_')[2])

# get alt
gtex['alt'] = gtex.variant_id.apply(lambda s: s.split('_')[3])

# get length ref
gtex['length_ref'] = gtex.ref.apply(lambda s: len(s))

# get end
gtex['end'] = gtex['start'] + gtex['length_ref'] - 1

# keep only columns annovar needs
df = gtex[['chrom', 'start', 'end', 'ref', 'alt']]
```

```python
# save df
df.to_csv('gtex_eqtls_annovar_input.tsv', sep='\t', index=False, header=None)
```
